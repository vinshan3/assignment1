```
# fibonacci series


def fib(index):
  if index <= 1:
    return(index)
  else:
    return(fib(index-1) + fib(index-2))
fib(7)
for i in range(10):
  print(fib(i))
```

```
# cross product

mat1 = np.array([[1,2],[3,4],[5,6]])
mat2 = np.array([[10,-4],[3,-7],[4,8]])
cross_mat = np.cross(mat1,mat2)
print(cross_mat)
```

```
# matrix creation

mat1 = np.random.randint(1,9, size = (2,2))
print(mat1)
mat2 = np.random.randint(-9,4, size = (2,2))
print(mat2)
```

```
# cross product

cross_prod = np.cross(mat1,mat2)
print(cross_prod)
```

```
# dot product

dot_prod = np.dot(mat1,mat2)
print(dot_prod)
```

```
# add_prod

add_prod = np.add(mat1,mat2)
print(add_prod)
```
